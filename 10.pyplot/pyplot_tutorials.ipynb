```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
apple_data = pd.read_csv('AAPL.csv')
```

```python
apple_data.head()
```

```python
apple_data.tail()
```

```python
dates = apple_data['Date']
opens = apple_data['Open']
highs = apple_data['High']
lows = apple_data['Low']
closes = apple_data['Close']
```

```python
plt.plot(dates, closes)
plt.show()
```

```python
plt.plot(dates, closes)
plt.plot(dates, opens)
plt.show()
```

```python
plt.plot(dates, highs)
plt.plot(dates, lows)
plt.show()
```

```python
fig = plt.figure()
plt.subplot(221)
plt.plot(dates, opens)

plt.subplot(222)
plt.plot(dates, closes)

plt.subplot(223)
plt.plot(dates, highs)

plt.subplot(224)
plt.plot(dates, lows)
plt.show()
```

```python
fig, axes = plt.subplots(4, 1)
axes[0].plot(dates, opens)
axes[1].plot(dates, closes)
axes[2].plot(dates, highs)
axes[3].plot(dates, lows)
plt.show()
```

```python
fig, axes = plt.subplots(2, 2)
axes[0, 0].plot(dates, opens)
axes[0, 1].plot(dates, closes)
axes[1, 0].plot(dates, highs)
axes[1, 1].plot(dates, lows)
plt.show()
```

```python
# Customising graphs 

plt.plot(dates, closes)
plt.title('Close prices of AAPL')
plt.xlabel('Dates')
plt.ylabel('Prices in USD')

plt.xticks(pd.date_range('2018-10-01', '2019-09-30', 5).strftime('%Y-%m-%d'))
plt.show()
```

```python
fig, axes = plt.subplots(1, 2)
fig.suptitle('Prices for AAPL', y=1.05)

axes[0].plot(dates, highs)
axes[0].set_title('High prices for AAPL')
axes[0].set_xlabel('Dates')
axes[0].set_ylabel('Prices in USD')
axes[0].set_xticks(pd.date_range('2018-10-01', '2019-09-30', 2).strftime('%Y-%m-%d'))

axes[1].plot(dates, lows)
axes[1].set_title('Low Prices for AAPL')
axes[1].set_xlabel('Dates')
axes[1].set_ylabel('Prices in USD')
axes[1].set_xticks(pd.date_range('2018-10-01', '2019-09-30', 2).strftime('%Y-%m-%d'))

plt.tight_layout()
plt.show()
```

```python
plt.plot(dates, highs)
plt.plot(dates, lows)

plt.title('Daily High and Low prices of AAPL')
plt.xlabel('Dates')
plt.ylabel('Prices in USD')

plt.xticks(pd.date_range('2018-10-01', '2019-09-30', 5).strftime('%Y-%m-%d'))
plt.legend(loc='upper center')
plt.show()
```

```python
plt.plot(dates, closes, 'g', lw=3)
plt.title('Close prices of AAPL')
plt.xlabel('Dates')
plt.ylabel('Prices in USD')

plt.xticks(pd.date_range('2018-10-01', '2019-09-30', 5).strftime('%Y-%m-%d'))
plt.show()
```
