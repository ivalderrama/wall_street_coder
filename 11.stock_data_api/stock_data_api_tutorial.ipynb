```python
import requests
from pprint import pprint
```

```python
def get_stock_data(stock_symbol, start_date, end_date):
    base_url = f"https://api.tiingo.com/tiingo/daily/{stock_symbol}/prices?"
    api_key = "a4e667e32d226a51c412a867bcd480676401844a"
    payload = {
        "token": api_key,
        "startDate": start_date,
        "endDate": end_date
    }
    
    response = requests.get(base_url, params=payload)
    
    return response
    
```

```python
def decode_data(data_response):
    json_response = data_response.json()
    
    dates = []
    highs = []
    lows = []
    
    for i in json_response:
        date = i['date']
        high = i['high']
        low = i['low']
        
        dates.append(date)
        highs.append(high)
        lows. append(low)
        
    return dates, highs, lows
```

```python
dates, highs, lows = decode_data(data_response)

print(dates)
print(highs)
print(lows)
```

```python
# Graphing Data

from matplotlib import pyplot as plt
```

```python
plt.plot(dates, highs, label='highs')
plt.plot(dates, lows, label='lows')
plt.title('High and Low Prices of AAPL')
plt.xlabel('Dates')
plt.ylabel('Prices in USD')
plt.legend(loc='upper left')
plt.show()
```
